```python
import pandas as pd
```

```python

melb_data = pd.read_csv('data/melb_data_fe.csv')
melb_df = melb_data.copy()
melb_df
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df.info()
```

```python
exclude_columns = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_uniq_vals = 150
for col in melb_df.columns:
    if col not in exclude_columns and melb_df[col].nunique() <= max_uniq_vals:
        melb_df[col] = melb_df[col].astype('category')
melb_df.info()
```

## 2. Сортировка данных в DataFrame
### Метод sort_values()

```python
melb_df.sort_values(by='Price', ascending=False).head(10)
```

```python
melb_df.sort_values(by='Date', ascending=False)
```

```python
melb_df.sort_values(by=['Distance', 'Price'], ascending=True).loc[::10, ['Distance', 'Price']]
```

##### Комбинирование сортировки с фильтрацией

```python
mask1 = melb_data['Type'] == 'townhouse'
mask2 = melb_data['AreaRatio'] < -0.8
mask3 = melb_data['SellerG'] == 'McGrath'
melb_data[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'], 
    ascending=[True, False],
    ignore_index=True
    ).loc[:, ['Date', 'AreaRatio']]
```

## 3. Группировка данных в DataFrame
### Метод groupby()
#### Группировка данных по одному критерию с одной агрегацией

```python
melb_df.groupby(by='Type').mean(numeric_only=True)
```

```python
melb_df.groupby(by='Type')['Price'].mean()
```

```python
melb_df.groupby(by='Regionname')['Distance'].min().sort_values(ascending=False)
```

#### Группировка данных по одному критерию с несколькими агрегациями - agg()

```python
melb_df.groupby(by='MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False)
```

```python
melb_df.groupby(by='MonthSale')['Price'].agg('describe').sort_values(by='count', ascending=False)
```

```python
melb_df.groupby('Regionname')['SellerG'].agg(
    ['nunique', set]
).sort_values(by='nunique', ascending=False)
```

## 4. Сводные таблицы
### Метод groupby как способ построения сводных таблиц

```python
melb_df.groupby('Rooms')[['Price', 'BuildingArea']].median()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()
```

### Метод pivot_table для построения сводных таблиц

```python
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

```python
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
).round()
```

```python
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    fill_value=0,
    aggfunc=['mean', 'median']
)
```

#### Многомерные сводные таблицы

```python
melb_df.pivot_table(
    values='Price',
    index=['Method', 'Type'],
    columns='Regionname',
    fill_value=0,
    aggfunc='median'
)
```

#### Доступ к данным в сводной таблице

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    fill_value=0,
    aggfunc=['mean', 'median']
)
pivot
```

```python
display(pivot.columns)
```

```python
display(pivot['mean']['house'])
```

## 5. Объединение DataFrame: знакомимся с новыми данными

```python
movies_data = pd.read_csv('data/movies.csv')
movies_df = movies_data.copy()
movies_df
```

```python
ratings1_data = pd.read_csv('data/ratings1.csv')
ratings1_df = ratings1_data.copy()
print(ratings1_df)
ratings2_data = pd.read_csv('data/ratings2.csv')
ratings2_df = ratings2_data.copy()
print(ratings2_df)
```

```python
dates_data = pd.read_csv('data/dates.csv')
dates_df = dates_data.copy()
dates_df
```

```python
print(movies_df['title'].nunique())
print(ratings1_df['userId'].nunique())
```

```python
dates_df['date'] = pd.to_datetime(dates_df['date'])
dates_df['date'].dt.year.value_counts()
```

## 6. Объединение DataFrame: concat

```python
ratings_df = pd.concat(
    [ratings1_df, ratings2_df],
    ignore_index=True
)
print(ratings_df.shape[0])
print(dates_df.shape[0])
```

```python
ratings_df = ratings_df.drop_duplicates(ignore_index=True)
ratings_df
```

```python
ratings_dates = pd.concat(
    [ratings_df, dates_df],
    axis=1
)
ratings_dates
```

## 7. Объединение DataFrame: join, merge
#### Типы объединений - inner, full/left/right outer 
#### Метод объединения join

```python
joined_false = ratings_dates.join(
    movies_df,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
joined = ratings_dates.join(
    movies_df.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined)
```

#### Метод объединения merge

```python
merged = ratings_dates.merge(
    movies_df,
    on='movieId',
    how='left'
)
merged.head(30)
```

```python
merged_ratings = ratings1_df.merge(ratings2_df, how='outer')
merged_ratings
```
