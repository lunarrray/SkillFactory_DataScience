### Feature Engineering

```python
import pandas as pd
melb_data = pd.read_csv('data/melb_data_ps.csv')
melb_data.head()
```

### Создание копии таблицы - copy()

```python
melb_df = melb_data.copy()
```

### Удаление столбцов  - drop()

```python
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

### Математические операции со столбцами - 

```python
total_rooms = melb_df['Bathroom'] + melb_df['Bedroom'] + melb_df['Rooms']
melb_df['MeanRoomsArea'] = melb_df['BuildingArea'] / total_rooms
melb_df.head()
```

```python
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_are = melb_data['BuildingArea'] + melb_data['Landsize']
melb_df['AreaRatio'] = diff_area / sum_are
melb_df.head()
```

## 4. Создание и преобразование столбцов с помощью функций

```python
melb_df['Address']
```

```python
def get_street_type(address: str) -> str:
    exclude_list = ['N', 'S', 'W', 'E']
    abbr_dict = {'Avenue': 'Av', 'Ave': 'Av', 'Boulevard': 'Bvd', 'Parade': 'Pde', 'Highway': 'Hvy', 'Esplanade': 'Esp', 'Grand': 'Grn'}
    address_list = address.split(' ')
    street_type = address_list[-1]
    if street_type in exclude_list:
        street_type = address_list[-2]
    if street_type in abbr_dict.keys():
        street_type = abbr_dict[street_type]
    return street_type
```

```python
street_types = melb_df['Address'].apply(get_street_type)
print(street_types.nunique())
```

```python
display(street_types.value_counts().tail(25))
display(street_types.value_counts().head(28))
```

```python
popular_stypes = street_types.value_counts().nlargest(10).index
popular_stypes
```

```python
melb_df['StreetType'] = street_types.apply(lambda x: x if x in popular_stypes else 'other')
melb_df['StreetType'].nunique()
```

```python
melb_df = melb_df.drop('Address', axis=1)
melb_df
```

## 5. Тип данных Category

### Признаки: категориальные и числовые
#### Категории в данных о недвижимости

```python
unique_list = []
for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtype)
    unique_list.append(item)
unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
).sort_values(by='Num_Unique', ignore_index=True)
print(unique_counts)
```

### Тип данных category

```python
display(melb_df.info())
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bathroom', 'Car', 'Bedroom']
max_unique_counts = 150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_counts and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```
